```python
from tkinter import *
from PIL import ImageTk, Image
import sqlite3


root = Tk()
root.title("Learn GUI Jun")
root.iconbitmap(r'Icons\PurpleFish.ico')
root.geometry("600x500")

# Databases



# Create function to delete a record

def clicked_delete_db():
    
    # Create a database or connect to one
    Myconn = sqlite3.connect('address_book.db')

    # Create cursor
    Mycursor = Myconn.cursor()

    # Delete a record
    
    Mycursor.execute(
    "DELETE from people WHERE oid = " + delete_box.get()
    )

    # Commit changes
    Myconn.commit()

    # Close Connection
    Myconn.close()




def clicked_create_db():
    
    # Create a database or connect to one
    Myconn = sqlite3.connect('address_book.db')

    # Create cursor
    Mycursor = Myconn.cursor()
    
    # Create table
    Mycursor.execute("""CREATE TABLE people (
        first_name text,
        last_name text,
        address text,
        city text,
        state text,
        zipcode integar
        )
                     """)

    # Commit changes
    Myconn.commit()

    # Close Connection
    Myconn.close()
    

def clicked_submit():
    # Create a database or connect to one
    Myconn = sqlite3.connect('address_book.db')

    # Create cursor
    Mycursor = Myconn.cursor()

    #Insert into Table
    Mycursor.execute(
        "INSERT INTO people VALUES (:f_name, :l_name, :address, :city, :state, :zipcode)",
        {
            'f_name': f_name.get(),
            'l_name': l_name.get(),
            'address': address.get(),
            'city': city.get(),
            'state': state.get(),
            'zipcode': zipcode.get()
        }         
                    ) 
    

    # Commit changes
    Myconn.commit()

    # Close Connection
    Myconn.close()

    # Clear text boxes
    f_name.delete(0, END)
    l_name.delete(0, END)
    address.delete(0, END)
    city.delete(0, END)
    state.delete(0, END)
    zipcode.delete(0, END)

    
    
# Create query function

def clicked_query():
    global query_label
    # Create a database or connect to one
    Myconn = sqlite3.connect('address_book.db')

    # Create cursor
    Mycursor = Myconn.cursor()
    
    # Query the database
    Mycursor.execute("SELECT *, oid FROM people")
    records = Mycursor.fetchall() #fetchmany(50) fetchone
    
    query_label.destroy()
    
    print_records = ""
    for list in records:
        print_records += str(list) + "\n"
        # Format anyway you want
        #print_records += str(list[0]) + " " str(list[1]) + "\n"
        
        
    query_label = Label(root,text=print_records)
    query_label.grid(row=9,column=0,columnspan=2)
    
    # Commit changes
    Myconn.commit()

    # Close Connection
    Myconn.close()


    
# Create edit function

def clicked_edit_db():
    editor = Tk()
    editor.title("Update Records")
    editor.iconbitmap(r'Icons\PurpleFish.ico')
    editor.geometry("600x500")
    
    # Create a database or connect to one
    Myconn = sqlite3.connect('address_book.db')

    # Create cursor
    Mycursor = Myconn.cursor()
    
    
    
    # Create Text Boxes

    f_name_editor = Entry(editor, width=30)
    f_name_editor.grid(row=0, column=1, padx=10, pady=(5,0))

    l_name_editor = Entry(editor, width=30)
    l_name_editor.grid(row=1, column=1, padx=10)

    address_editor = Entry(editor, width=30)
    address_editor.grid(row=2, column=1, padx=10)

    city_editor = Entry(editor, width=30)
    city_editor.grid(row=3, column=1, padx=10)

    state_editor = Entry(editor, width=30)
    state_editor.grid(row=4, column=1, padx=10)

    zipcode_editor = Entry(editor, width=30)
    zipcode_editor.grid(row=5, column=1, padx=10)


    # Query the database
    Mycursor.execute("SELECT *, oid FROM people")
    records = Mycursor.fetchall() #fetchmany(50) fetchone

    for items in records:
        f_name_editor.insert(0,items[0])
        l_name_editor.insert(0,items[1])
        address_editor.insert(0,items[2])
        city_editor.insert(0,items[3])
        state_editor.insert(0,items[4])
        zipcode_editor.insert(0,items[5])

    # Create Text Labels
    f_name_label_editor = Label(editor, text="First Name:")
    f_name_label_editor.grid(row=0,column=0,pady=(5,0))

    l_name_label_editor = Label(editor, text="Last Name:")
    l_name_label_editor.grid(row=1,column=0)

    address_label_editor = Label(editor, text="Address:")
    address_label_editor.grid(row=2,column=0)

    city_label_editor = Label(editor, text="City:")
    city_label_editor.grid(row=3,column=0)

    state_label_editor = Label(editor, text="State:")
    state_label_editor.grid(row=4,column=0)

    zipcode_label_editor = Label(editor, text="Zipcode:")
    zipcode_label_editor.grid(row=5,column=0)


    # Create save buttons

    save_btn = Button(editor, text="Save Records", command=clicked_save)
    save_btn.grid(row=6, column=0, columnspan=2,pady=3,padx=10,ipadx=127)
    
    
    # Commit changes
    Myconn.commit()

    # Close Connection
    Myconn.close()
    
    

    
    
# Create Text Boxes

f_name = Entry(root, width=30)
f_name.grid(row=0, column=1, padx=10, pady=(5,0))

l_name = Entry(root, width=30)
l_name.grid(row=1, column=1, padx=10)

address = Entry(root, width=30)
address.grid(row=2, column=1, padx=10)

city = Entry(root, width=30)
city.grid(row=3, column=1, padx=10)

state = Entry(root, width=30)
state.grid(row=4, column=1, padx=10)

zipcode = Entry(root, width=30)
zipcode.grid(row=5, column=1, padx=10)

delete_box = Entry(root, width=30)
delete_box.grid(row=10,column=1, padx=10)


# Create Text Labels
f_name_label = Label(root, text="First Name:")
f_name_label.grid(row=0,column=0,pady=(5,0))

l_name_label = Label(root, text="Last Name:")
l_name_label.grid(row=1,column=0)

address_label = Label(root, text="Address:")
address_label.grid(row=2,column=0)

city_label = Label(root, text="City:")
city_label.grid(row=3,column=0)

state_label = Label(root, text="State:")
state_label.grid(row=4,column=0)

zipcode_label = Label(root, text="Zipcode:")
zipcode_label.grid(row=5,column=0)

query_label = Label(root,text="#######################\nHere will be records\n#######################")
query_label.grid(row=9,column=0,columnspan=2)

delete_box_label = Label(root, text="ID Number for Modify:")
delete_box_label.grid(row=10,column=0)


# Create database file

create_db_btn = Button(root, text= "Create Database File", command=clicked_create_db)
create_db_btn.grid(row=6, column=0, columnspan=2,pady=10,padx=10,ipadx=95)


# Create submite buttons

submit_btn = Button(root, text= "Add Record To Database", command=clicked_submit)
submit_btn.grid(row=7, column=0, columnspan=2,pady=3,padx=10,ipadx=95)


# Create a query buttons

query_btn = Button(root, text="Show Records", command=clicked_query)
query_btn.grid(row=8, column=0, columnspan=2,pady=3,padx=10,ipadx=123)

# Create a delete query buttons

delete_btn = Button(root, text="Delete Record", command=clicked_delete_db)
delete_btn.grid(row=11, column=0, columnspan=2,pady=3,padx=10,ipadx=122)

# Create edit buttons

edit_btn = Button(root, text="Edit Records", command=clicked_edit_db)
edit_btn.grid(row=12, column=0, columnspan=2,pady=3,padx=10,ipadx=127)





mainloop()
```
